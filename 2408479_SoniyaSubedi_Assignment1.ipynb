```python
#Task 3
#Qno1
#Qno2
q1_gdp, q3_gdp=south_asian_df['Log GDP per capita'].quantile([0.25,0.75])
iqr_gdp=q3_gdp - q1_gdp
lower_gdp=q1_gdp - 1.5 * iqr_gdp
upper_gdp=q3_gdp + 1.5 * iqr_gdp
q1_score, q3_score = south_asian_df['score'].quantile([0.25, 0.75])
iqr_score=q3_score - q1_gdp
lower_score=q1_score - 1.5*iqr_score
upper_score=q3_score + 1.5*iqr_score
outliers=south_asian_df[(south_asian_df['Log GDP per capita']<lower_gdp)| (south_asian_df['Log GDP per capita']>upper_gdp)|(south_asian_df['score']<lower_score)|(south_asian_df['score']>upper_score)]
print(f"Outlier Countries in South asia:\n",outliers)
```

```python
#Qno3
plt.figure(figsize=(10,6))
sns.scatterplot(x='Log GDP per capita',y='score',data=south_asian_df,label='Non-Outliers')
sns.scatterplot(x='Log GDP per capita',y='score',data=outliers,color='brown',label='Outliers')
plt.title('Scatter PLot of Log GDP per capita Vs Score')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
#Qno6
def detect_outliers(df,column):
  q1, q3=df[column].quantile([0.25, 0.75])
  iqr=q3 - q1
  lower_bound= q1 - 1.5 * iqr
  upper_bound= q3 + 1.5 * iqr
  return df[(df[column]< lower_bound) | (df[column]>upper_bound)]

outliers_sac=detect_outliers(south_asian_df, 'score')
outliers_mec=detect_outliers(middle_eastern_df, 'score')

print(f"South Asian Outliers:", outliers_sac)
print(f"Middle Eastern Outliers:", outliers_mec)
```

```python
#Qno6.2
plt.figure(figsize=(10,6))
sns.scatterplot(x ='Log GDP per capita', y= 'score', data = south_asian_df, label='South Asian')
sns.scatterplot(x ='Log GDP per capita', y= 'score', data=middle_eastern_df, label='Middle Eastern')
sns.scatterplot(x ='Log GDP per capita', y= 'score', data=outliers_sac, color = 'brown', label='South Asian Outliers')
sns.scatterplot(x ='Log GDP per capita', y= 'score', data=outliers_mec, color= 'green', label='Middle Eastern Outliers')
plt.title("Outliers in South Asian and Middle Eastern")
plt.legend()
plt.show()
```
